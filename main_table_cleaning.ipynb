```python
import pandas as pd
from scipy import stats
```

```python
main_table = pd.read_csv('data/main_table.csv')
```

```python
main_table
```

```python
def nulls_to_zeros(value):
    if pd.isna(value):
        return 0
    else:
        return value
```

```python
for index, row in main_table.iterrows():
    main_table.loc[index, 'class_d_players'] = nulls_to_zeros(row.class_d_players)
    main_table.loc[index, 'class_c_players'] = nulls_to_zeros(row.class_c_players)
    main_table.loc[index, 'class_b_players'] = nulls_to_zeros(row.class_b_players)
    main_table.loc[index, 'class_a_players'] = nulls_to_zeros(row.class_a_players)
    main_table.loc[index, 'candidate_masters'] = nulls_to_zeros(row.candidate_masters)
    main_table.loc[index, 'national_masters'] = nulls_to_zeros(row.national_masters)
    main_table.loc[index, 'fide_masters'] = nulls_to_zeros(row.fide_masters)
    main_table.loc[index, 'international_masters'] = nulls_to_zeros(row.international_masters)
    main_table.loc[index, 'grandmasters'] = nulls_to_zeros(row.grandmasters)
```

```python
main_table['tru_hr_players'] = main_table['candidate_masters'] + main_table['national_masters'] +  main_table['international_masters'] + main_table['fide_masters'] + main_table['grandmasters']
```

```python
main_table['tru_hr_rank'] = main_table['tru_hr_players'].rank(ascending=False)
```

```python
main_table.to_csv('data/upd_main_table.csv')
```

```python
main_table['pop'].corr(main_table['tru_hr_players'])
```
