```python
import pandas as pd
import re
import requests
from bs4 import BeautifulSoup as BS
import seaborn as sns
import os
```

```python
world_edu_db = BS(requests.get("https://worldtop20.org/education-database/").text)
```

```python
country_edu_ranking = []

for i in world_edu_db.find_all("p")[4:-1]:
    country_edu_ranking.append(i.text)
```

```python
# The 1st country was formatted a bit differently, I just manually inputted it.

country_edu_ranking.insert(0, "1. Denmark")
```

```python
# also got some junk so I culled them by running this twice

country_edu_ranking.pop(100)
```

```python
country_edu_ranking
```

```python
world_edu_rank = pd.DataFrame()
```

```python
country_edu_ranking_cleaner = []

for i in country_edu_ranking:
    country_edu_ranking_cleaner.append(i.split(". ")[1])
```

```python
world_edu_rank['country'] = country_edu_ranking_cleaner
```

```python
world_edu_rank = world_edu_rank.reset_index()
```

```python
world_edu_rank['index'] = world_edu_rank['index']+1
```

```python
world_edu_rank = world_edu_rank.rename(columns={"index":"rank"})
```

```python
world_edu_rank
```

```python
world_edu_rank.to_csv("data/world_edu_rank.csv", index=False)
```

also scraped the country abbreviations here cause it didn't need a whole notebook

```python
country_codes_site = BS(requests.get("https://www.iban.com/country-codes").text)
```

```python
full_country_names = country_codes_site.find("table").tbody.find_all('td')[0::4]
```

```python
country_alpha3s = country_codes_site.find("table").tbody.find_all('td')[2::4]
```

```python
country_codes = pd.DataFrame()
```

```python
country_codes['country'] = full_country_names
country_codes['a3'] = country_alpha3s
```

```python
country_codes #whoops forgot to extract the text
```

```python
def remove_html_tags(tag):
    return tag.text
```

```python
for index, row in country_codes.iterrows():
    country_codes.loc[index, 'country'] = remove_html_tags(row.country)
    country_codes.loc[index, 'a3'] = remove_html_tags(row.a3)
```

```python
country_codes
```

```python
country_codes.to_csv("data/country_codes.csv", index=False)
```
